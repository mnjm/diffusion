```python
import numpy as np
import matplotlib.pyplot as plt
from ddpm import Diffusion
from omegaconf import OmegaConf
from utils import torch_get_device
```

```python
linear_cfg_yml = """
timesteps: 1000
beta_start: 1e-4
beta_end: 0.02
schedule: linear
cfg: # classifier free guidance
    enable: false
    scale: 3.0
"""

cosine_cfg_yml = """
timesteps: 256
beta_start: 1e-4
beta_end: 0.02
schedule: cosine
cfg: # classifier free guidance
    enable: false
    scale: 3.0
"""
```

```python
diffusion = Diffusion(OmegaConf.create(linear_cfg_yml))
plt.figure(figsize=(20, 4))
plt.subplot(1, 4, 1)
plt.plot(diffusion.betas.cpu().numpy())
plt.title("Betas")
plt.subplot(1, 4, 2)
plt.plot(diffusion.alphas_bar.cpu().numpy())
plt.title("Alpha_bar")
plt.subplot(1, 4, 3)
plt.plot(diffusion.sqrt_1_minus_alphas_bar.cpu().numpy())
plt.title("sqrt(1 - alpha_bar)")
plt.subplot(1, 4, 4)
plt.plot((diffusion.alphas**-0.5).cpu().numpy())
plt.title("1/sqrt(alpha)")
plt.show()
```
